```python
# Dependencies and Setup
import pandas as pd

# File to Load
exact_loc_csv = "lat_and_long.csv"
main_csv = "resources/red_and_white_combined.csv"

# Read lat_long data file and store into Pandas DataFrames
loc_df = pd.read_csv(exact_loc_csv)
df = pd.read_csv(main_csv)
```

```python
# delete a single row by index value 0
df = df.drop(labels=0, axis=0)
df = df.reset_index()
df.head()
```

```python
df = df.drop(columns="index")
df.head()
```

```python
df = df.rename(columns={"Name": "name", "Country": "country", "Region": "region", "Winery": "winery", "Rating": "rating", "NumberOfRatings": "number_of_ratings", "Price": "price", "Year": "year", "Type": "type"})
df.head()
```

```python
loc_df = loc_df.drop(columns=["name", "country"])
loc_df.head()
```

```python
new_df = pd.merge(df, loc_df, how="right", on="winery")
new_df
```
